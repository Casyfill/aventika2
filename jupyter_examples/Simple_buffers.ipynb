```python
import geopandas as gp
import pandas as pd
```

```python
from code.main import getSettings, data_preload
```

```python
settings = getSettings()

banks_path = settings['data_path'] + settings['files']['banks']

banks = gp.read_file(banks_path)[['geometry', 'office_id']].to_crs(epsg='32637')
```

```python
# banks
banks.head(100).plot()
```

```python
walk = banks.buffer(1000, resolution=1)
step = banks.buffer(400, resolution=1)
```

```python
walkdf = banks[['office_id', 'geometry']]
walkdf['geometry'] = walk

stepdf = banks[['office_id', 'geometry']]
stepdf['geometry'] = step
```

```python
walkdf['geometry'] =  walkdf['geometry'].difference(step)
walkdf.head(10).plot()
```

```python
walkdf['type'] = 'foot'
stepdf['type'] = 'stepless'
```

```python
buffers = pd.concat([walkdf, stepdf])
```

```python
buffers.head(2)
```

```python
buffers.to_crs(espg=, inplace1)
```

```python
with open(settings['data_path'] + 'circles_simple.geojson','w') as f:
    f.write(buffers.to_crs(epsg=4326).to_json())
```
